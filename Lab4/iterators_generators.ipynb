```python
def gen_number(x):
    for i in range(1, x + 1):
        yield i ** 2


n = int(input())
gen_generated = gen_number(n)

print(f"Square numbers before {n}")
for j in gen_generated:
    print(j)
```

```python
def gen_even(n):
    for i in range(0, n + 1):
        if i % 2 == 0:
            yield i
    

n = int(input())
even_numbers = gen_even(n)

print("Even numbers before:", n)

for i in even_numbers:
    print(i)
```

```python
def gen_iter(n):
    for i in range(n):
        if i % 3 == 0 and i % 4 == 0:
            yield i


n = int(input())
numbers = gen_iter(n)
print("That divisible on 3 and 4:")
for i in gen_iter(n):
    print(i)

```

```python
def square(a, b):
    for i in range(a, b + 1):
        yield  i ** 2


a, b = int(input()), int(input())
number_square = square(a, b)
print("Squares:")
for i in number_square:
    print(i)
```

```python
def reverse(n):
    for i in range(n, 0, -1):
        yield i


n = int(input())
reversoo = reverse(n)

print("Reverse number")

for i in reversoo:
    print(i)
```
